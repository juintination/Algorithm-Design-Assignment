```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class LCS:
    
    def __init__(self, x, y):
        self.count = 0
        self.x, self.y = x, y
        self.lx, self.ly = len(x) + 1, len(y) + 1
        self.dp = np.array([np.NaN] * self.lx * self.ly).reshape(self.lx, self.ly)
        
    # 각 lcs의 계산 횟수를 구한 후에 다시 self.count를 0으로 초기화하는 메서드
    def initCount(self):
        self.count = 0
        
    def initDP(self):
        self.dp = np.array([np.NaN] * self.lx * self.ly).reshape(self.lx, self.ly)
        # self.dp = [[np.NaN for _ in range(n + 1)] for _ in range(n + 1)]
        
    # 계산이 발생할 때 계산 횟수만큼 self.count에 더하는 메서드
    def plusCount(self, count):
        self.count += count
        
    # 각 lcs 알고리즘이 끝날 때마다 구한 계산 횟수를 return하고 초기화하는 메서드
    def popCount(self):
        rv = self.count # return 값
        self.initCount() # self.count 초기화
        return rv
    
    def setXY(self, x, y):
        self.initCount()
        self.x, self.y = x, y
        self.lx, self.ly = len(x) + 1, len(y) + 1
        self.dp = np.array([np.NaN] * self.lx * self.ly).reshape(self.lx, self.ly)
    
    def rLCS(self, i, j):
        self.plusCount(1) #
        if i == 0 or j == 0:
            return 0
        elif self.x[i - 1] == self.y[j - 1]:
            return self.rLCS(i - 1, j - 1) + 1
        else:
            return max(self.rLCS(i - 1, j), self.rLCS(i, j - 1))
        
    def rmLCS(self, i, j):
        self.plusCount(1) #
        if i == 0 or j == 0:
            return 0
        if np.isnan(self.dp[i][j]):
            if self.x[i - 1] == self.y[j - 1]:
                self.dp[i][j] = self.rmLCS(i - 1, j - 1) + 1
            else:
                self.dp[i][j] = max(self.rmLCS(i - 1, j), self.rmLCS(i, j - 1))
        return int(self.dp[i][j])
    
    def dpLCS(self):
        self.initDP()
        for i in range(self.lx):
            self.dp[i][0] = 0
        for j in range(self.ly):
            self.dp[0][j] = 0
        for i in range(1, self.lx):
            for j in range(1, self.ly):
                self.plusCount(1) #
                if self.x[i - 1] == self.y[j - 1]:
                    self.dp[i][j] = int(self.dp[i - 1][j - 1]) + 1
                else:
                    self.dp[i][j] = max([self.dp[i - 1][j], self.dp[i][j - 1]])
        return int(self.dp[len(self.x)][len(self.y)])
    
    def pathLCS(self):
        stack = []
        i, j = len(self.x), len(self.y)
        # print(self.dp) # for test
        if self.dp[i][j] == 0:
            return ""
        while i > 0 and j > 0:
            self.plusCount(1) #
            if self.dp[i][j] == self.dp[i - 1][j]:
                i -= 1
            elif self.dp[i][j] == self.dp[i][j - 1]:
                j -= 1
            else:
                stack.append(self.x[i - 1])
                i, j = i - 1, j - 1
        stack.reverse()
        return "".join(stack)
    
    def dpLCSforConsecutiveSubstring(self):
        maxLen, endPos = 0, (0, 0)
        for i in range(self.lx):
            self.dp[i][0] = 0
        for j in range(self.ly):
            self.dp[0][j] = 0
        for i in range(1, self.lx):
            for j in range(1, self.ly):
                self.plusCount(1) #
                if self.x[i - 1] == self.y[j - 1]:
                    self.dp[i][j] = int(self.dp[i - 1][j - 1]) + 1
                    if self.dp[i][j] > maxLen:
                        maxLen = self.dp[i][j]
                        endPos = (i, j)
                else:
                    self.dp[i][j] = 0
        return endPos

    def pathLCSforLongestConsecutiveSubstring(self):
        endPos = self.dpLCSforConsecutiveSubstring()
        i, j = endPos
        stack = []
        while self.dp[i][j] > 0:
            self.plusCount(1) #
            stack.append(self.x[i - 1])
            i, j = i - 1, j - 1
        stack.reverse()
        return "".join(stack)
```

```python
print('LCS 알고리즘이 정상적으로 잘 동작하는지 확인하기 위한 테스트 1')

x1, y1 = 'abcde', 'bde'
print('x1:', x1, '/ y1:', y1)

lcsForTest = LCS(x1, y1)

print()

# rLCS Test
print('rLCS:', lcsForTest.rLCS(len(x1), len(y1)))
print('count:', lcsForTest.popCount(), '\n')

# rmLCS Test
print('rmLCS:', lcsForTest.rmLCS(len(x1), len(y1)))
print('count:', lcsForTest.popCount(), '\n')

# dpLCS Test
print('dpLCS:', lcsForTest.dpLCS())
print('count:', lcsForTest.popCount(), '\n')

# pathLCS Test
print('pathLCS:', lcsForTest.pathLCS())
print('count:', lcsForTest.popCount(), '\n')

# getConsecutiveSubstring Test
print('consecutiveSubstring:', lcsForTest.pathLCSforLongestConsecutiveSubstring())
print('count:', lcsForTest.popCount())
```

```python
dataSet = {}
inStrs = [["".join(chr(random.randint(ord('a'), ord('z'))) for _ in range(10)) for _ in range(2)] \
          for _ in range(10)]
inStrs = [(inStr[0], inStr[1]) for inStr in inStrs]

for i in range(len(inStrs)):
    dataSet[i] = inStrs[i]
    x, y = dataSet[i]
    print(f'x{i}: {x} / y{i}: {y}')
```

```python
operateCounts = {'x':[], 'y':[], 'rLCS':[], 'rmLCS':[], 'dpLCS':[], 'longestCommonSubstring':[], 'pathLCS':[]}
lcs = LCS("", "")

for inStr in inStrs:
    x, y = inStr
    lcs.setXY(x, y)
    
    operateCounts['x'].append(x)
    operateCounts['y'].append(y)
    lcs.rLCS(len(x), len(y))
    operateCounts['rLCS'].append(lcs.popCount())
    lcs.rmLCS(len(x), len(y))
    operateCounts['rmLCS'].append(lcs.popCount())
    lcs.dpLCS()
    operateCounts['dpLCS'].append(lcs.popCount())
    operateCounts['longestCommonSubstring'].append(lcs.pathLCS())
    operateCounts['pathLCS'].append(lcs.popCount())
```

```python
print('입력 문자열에 대한 각 lcs의 연산 횟수')

df = pd.DataFrame(operateCounts)
display(df)
```

```python
xticklabels = [f'x: {x} / y: {y}' for x, y in inStrs]

lcsAlgorithms = [['rLCS', 'rmLCS'],
                ['dpLCS', 'pathLCS']]

barColors = [['red', 'blue'],
               ['green', 'orange']]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
fig.suptitle('각 lcs 알고리즘의 연산 횟수')

for i in range(2):
    for j in range(2):
        lcsAlgorithm = lcsAlgorithms[i][j]
        barColor = barColors[i][j]
        axes[i][j].bar(range(len(inStrs)), df[lcsAlgorithm], color=barColor)
        axes[i][j].set_title(lcsAlgorithm)
        axes[i][j].set_xticks(np.arange(10))
        axes[i][j].set_xticklabels(xticklabels, rotation=90)
        
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 10))
fig.suptitle('각 알고리즘의 연산 횟수')

lcsAlgorithms = ['rLCS', 'rmLCS', 'dpLCS', 'pathLCS']
barColors = ['red', 'blue', 'green', 'orange']
axes = axes.flatten()

for axIdx, inStr in enumerate(dataSet.values()):
    x, y = inStr
    xticklabels = [f'{algorithm}: {val}' for algorithm, val in zip(lcsAlgorithms, df.loc[axIdx, lcsAlgorithms])]

    bars = axes[axIdx].bar(lcsAlgorithms, df.loc[axIdx, lcsAlgorithms], color=barColors)
    axes[axIdx].set_title(f'x: {x} / y: {y}')
    axes[axIdx].set_xlabel('')
    axes[axIdx].set_xticks(np.arange(len(lcsAlgorithms)))
    axes[axIdx].set_xticklabels(xticklabels, rotation=90)
    axes[axIdx].legend(bars, lcsAlgorithms, title='Algorithm')

plt.tight_layout()
plt.show()
```

```python
additionalCounts = {'x':[], 'y':[], 'longestConsecutiveSubstring':[], 'pathLCSforLongestConsecutiveSubstring':[]}

for inStr in inStrs:
    x, y = inStr
    lcs.setXY(x, y)
    
    additionalCounts['x'].append(x)
    additionalCounts['y'].append(y)
    additionalCounts['longestConsecutiveSubstring'].append(lcs.pathLCSforLongestConsecutiveSubstring())
    additionalCounts['pathLCSforLongestConsecutiveSubstring'].append(lcs.popCount())
```
